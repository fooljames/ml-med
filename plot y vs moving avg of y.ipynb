```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import time
from utils import *
import seaborn as sns
%matplotlib inline

path = 'D:/03 Project/01 KKU MLMED'
```

```python
file_list = glob.glob(path + "/phlilps*.csv")
phlilp = pd.DataFrame()
list_ = []
for file_ in file_list:
    if file_ == file_list[0]:
        df = pd.read_csv(file_, low_memory=False, index_col=None )
    else:
        df = pd.read_csv(file_, low_memory=False, index_col=None , header = 0)
    phlilp = phlilp.append(df)


phlilp['dataset_y_na_flag'] = np.where(phlilp['dataset_a0188'].isnull() == True, '1','0')
# nms[nms.name.notnull()]
# phlilp = phlilp.loc[phlilp['dataset_y_na_flag'] == '0']

phlilp['dataset_datetime'] = pd.to_datetime(phlilp['dataset_datetime'], format = '%Y%m%d%H%M%S')
phlilp = phlilp.sort_values(by = ['dataset_location','dataset_datetime'])
phlilp['dataset_datetime_diff'] = phlilp.groupby('dataset_location')['dataset_datetime'].diff()
```

```python
def moving_avg(df):
    df.index = df['dataset_datetime']
    f = lambda x: x.rolling('600s').mean()
    df['dataset_moving_avg'] = df.groupby('dataset_location')['dataset_a0188'].apply(f)
    df.index = range(len(df))
    return df
```

```python
moving_avg(phlilp)
```

```python
df = phlilp.loc[:,('dataset_location', 'dataset_datetime', 'dataset_a0188', 'dataset_moving_avg')]
df['SpO2'] = df['dataset_a0188']
df['SpO2_MovingAvg'] = df['dataset_moving_avg']
```

```python
import plotly as py
import cufflinks as cf
from plotly.offline import plot
from plotly.graph_objs import * 
from plotly import tools

fig = tools.make_subplots(rows=5, cols=3, shared_xaxes=True, subplot_titles=df.dataset_location.unique())

for i, locat in enumerate(df.dataset_location.unique()):
    df_locat = df[df.dataset_location==locat].copy()
    for col in ['SpO2', 'SpO2_MovingAvg']:
        r = int(np.floor(i/3)+1)
        c = (i%3)+1
        fig.append_trace({'x': df_locat.dataset_datetime, 'y': df_locat[col], 'type': 'scatter',  'name': col, 'legendgroup': locat}, r, c, )
    
        
plot(fig, filename='my-graph.html')
```
